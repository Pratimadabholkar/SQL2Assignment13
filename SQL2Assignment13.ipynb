```python
#Read the following data set:
#https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data
#Task:
#1. Create an sqlalchemy engine using a sample from the data set

#import libraries

import sqlalchemy
import pandas as pd
from sqlalchemy import create_engine

#read csv file in DataFrame and export into sqladult Table
#Database: myData.db
#Table Name: sqladult

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',sep=",",header=None)
df.columns=['age','workclass','fnlwgt','education','educationNum','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']
engine = create_engine('sqlite:///myData.db')
df.to_sql('sqladult', engine)
engine.execute("SELECT * FROM sqladult").fetchall()
```

```python
# Display 10 records from the table sqladult Table

import sqlite3 as lite
import sys

con = lite.connect('myData.db')

with con:    
    cur = con.cursor()    
    cur.execute("SELECT * FROM sqladult LIMIT 10;")
    rows = cur.fetchall()
    for row in rows:
        print(row)
```

```python
#1. Update Query
# Update 12th education to 12std for all records
# code for update operation 
import sqlite3 
  
# database name to be passed as parameter 
conn = sqlite3.connect('myData.db') 
  
# update the records 
conn.execute("UPDATE sqladult set education = '12std' WHERE education = ' 12th'") 
conn.commit() 
  
print("Total number of rows updated :", conn.total_changes)
  
cursor = conn.execute("SELECT distinct(education) FROM sqladult") 
for row in cursor: 
   print(row) 
  
conn.close()
```

```python
#2. Update Query
# Update workclass '?' with 'NA
# code for update operation 
import sqlite3 
  
# database name to be passed as parameter 
conn = sqlite3.connect('myData.db') 
  
# update the records 
conn.execute("UPDATE sqladult set workclass = 'NA' WHERE workclass = ' ?'") 
conn.commit() 
  
print("Total number of rows updated :", conn.total_changes)
  
cursor = conn.execute("SELECT distinct(workclass) FROM sqladult") 
for row in cursor: 
   print(row) 
  
conn.close()
```

```python
#3. Write two delete queries
# First query - delete records where age = 50
import sqlite3 
  
# database name to be passed as parameter 
conn = sqlite3.connect('myData.db') 
  
# update the records 
conn.execute("DELETE from sqladult WHERE age = 50") 
conn.commit() 
  
print("Total number of rows Deleted :", conn.total_changes)
  
cursor = conn.execute("SELECT * from sqladult where age = 50") 
for row in cursor: 
   print(row) 
  
conn.close()
```
